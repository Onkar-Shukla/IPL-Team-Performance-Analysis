Analyze team performance based on the winning percentages calculated from the `df_historical` and `df_ball_by_ball` datasets.

## Load and prepare data

### Subtask:
Load the `df_historical` and `df_ball_by_ball` dataframes and handle any potential data type issues, particularly in `df_ball_by_ball`.


**Reasoning**:
Load the specified CSV files into pandas DataFrames and display their head and info to verify the loading and data types.



```python
df_historical = pd.read_csv('ipl_historical.csv')
df_ball_by_ball = pd.read_csv('ipl_ball_by_ball_data.csv.zip', dtype={'10': str, '11': str, '12': str}, low_memory=False)

print("df_historical head:")
display(df_historical.head())
print("\ndf_historical info:")
df_historical.info()

print("\ndf_ball_by_ball head:")
display(df_ball_by_ball.head())
print("\ndf_ball_by_ball info:")
df_ball_by_ball.info()
```

## Identify match outcomes

### Subtask:
Determine the winning team for each match from the `df_historical` dataframe.


**Reasoning**:
Extract the winning team from the 'match_result_text' column and handle cases with no clear winner.



```python
# Extract the winning team from the 'match_result_text' column
df_historical['winning_team'] = df_historical['match_result_text'].apply(
    lambda x: x.split(' won by')[0].strip() if 'won by' in str(x) else ('No Result' if 'no result' in str(x).lower() else 'Tie')
)

# Display the first few rows with the new column
display(df_historical[['match_result_text', 'winning_team']].head())
```

## Calculate team wins

### Subtask:
Count the number of wins for each team.


**Reasoning**:
Count the number of wins for each team using the 'winning_team' column in df_historical, reset the index, and rename the columns.



```python
team_wins = df_historical.groupby('winning_team').size().reset_index(name='wins')
team_wins = team_wins.rename(columns={'winning_team': 'team'})
display(team_wins)
```

## Calculate total matches played

### Subtask:
Determine the total number of matches played by each team.


**Reasoning**:
Calculate the total matches played by each team by counting their appearances in 'team1' and 'team2' columns and summing them.



```python
team1_counts = df_historical['team1_name'].value_counts().reset_index()
team1_counts.columns = ['team', 'team1_matches']

team2_counts = df_historical['team2_name'].value_counts().reset_index()
team2_counts.columns = ['team', 'team2_matches']

total_matches = pd.merge(team1_counts, team2_counts, on='team', how='outer').fillna(0)
total_matches['total_matches'] = total_matches['team1_matches'] + total_matches['team2_matches']

total_matches = total_matches[['team', 'total_matches']]

display(total_matches)
```

## Calculate winning percentage

### Subtask:
Calculate the winning percentage for each team using the number of wins and total matches played.

**Reasoning**:
Merge the `team_wins` and `total_matches` dataframes on the 'team' column and calculate the winning percentage. Handle potential division by zero for teams with no matches played.

```python
import numpy as np

# Merge team_wins and total_matches
team_performance = pd.merge(team_wins, total_matches, on='team', how='left').fillna(0)

# Calculate winning percentage, handling potential division by zero
team_performance['winning_percentage'] = (team_performance['wins'] / team_performance['total_matches']) * 100
team_performance['winning_percentage'] = team_performance['winning_percentage'].replace([np.inf, -np.inf], 0) # Replace inf with 0

# Display the result
display(team_performance[['team', 'wins', 'total_matches', 'winning_percentage']].sort_values(by='winning_percentage', ascending=False))
```

# Task
Analyze team performance by calculating and visualizing the average scores of each team based on the data in "cricket_matches.csv".

## Extract team scores

### Subtask:
Get the runs scored by each team from `df_historical`.


**Reasoning**:
Select and rename columns from df_historical for both team1 and team2, then concatenate the results into a single DataFrame to get the runs scored by each team.



```python
team1_scores = df_historical[['team1_name', 'team1_runs_scored']].copy()
team1_scores.columns = ['team', 'runs_scored']

team2_scores = df_historical[['team2_name', 'team2_runs_scored']].copy()
team2_scores.columns = ['team', 'runs_scored']

all_team_scores = pd.concat([team1_scores, team2_scores], ignore_index=True)

display(all_team_scores.head())
```

## Calculate average scores

### Subtask:
Calculate the average runs scored for each team.


**Reasoning**:
Calculate the average runs scored for each team by grouping the `all_team_scores` DataFrame by team and calculating the mean of 'runs_scored', then reset the index and rename the column.



```python
average_team_scores = all_team_scores.groupby('team')['runs_scored'].mean().reset_index()
average_team_scores = average_team_scores.rename(columns={'runs_scored': 'average_runs_scored'})
display(average_team_scores)
```

## Display average scores

### Subtask:
Show the calculated average scores for each team.


**Reasoning**:
Display the average team scores, sorted by the average runs scored in descending order.



```python
average_team_scores_sorted = average_team_scores.sort_values(by='average_runs_scored', ascending=False)
display(average_team_scores_sorted)
```

```python
# Sort for better visualization
average_team_scores_sorted = average_team_scores.sort_values(by='average_runs_scored', ascending=True)

plt.figure(figsize=(12, 8))
plt.barh(average_team_scores_sorted['team'], average_team_scores_sorted['average_runs_scored'], color='skyblue')
plt.xlabel('Average Runs Scored')
plt.title('Average Runs Scored per Team in IPL')
plt.tight_layout() # Adjust layout to prevent labels overlapping
plt.show()
```
